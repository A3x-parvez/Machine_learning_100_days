```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("jharkhand-data-set.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
plt.scatterplot()
```

```python
df_june = df[df['Month'] == 'june']
```

```python
df_june
```

```python
# Assuming the two columns are named 'Column1' and 'Column2'
x = df_june['Water level']
y = df_june['Discharge']

plt.figure(figsize=(10, 10))
plt.plot(x, y, marker='o', linestyle='-', color='blue', label='Connection')
plt.title('Connection Between Water level and Discharge in june')
plt.xlabel('Water level')
plt.ylabel('Discharge')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 10))

# Plot for Column1
sns.histplot(x, kde=True, color='blue', label='Column1', bins=10)

# Plot for Column2
sns.histplot(y, kde=True, color='green', label='Column2', bins=10)

# Add labels, legend, and title
plt.title('Histogram of Column1 and Discharge')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', marker='o', label='Water Level vs Discharge')

# Add titles and labels
plt.title('Relationship Between Water Level and Discharge')
plt.xlabel('Water Level')
plt.ylabel('Discharge')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
