# gather data from all pages

```python
for i in range(1,501):
    movie_json = Fetch_pagedata(i) 
    raw_temp_df = pd.DataFrame(movie_json['results'])
    temp_df = raw_temp_df[['id','title','release_date','overview','popularity','vote_average','vote_count','poster_path']]
    movies = pd.concat([movies, temp_df], ignore_index=True)
    print("page ",i," fetched successfully")
```
